```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

# Load the dataset
print('Loading dataset...')
df = pd.read_parquet('data.parquet')

# Remove null values
print('Removing null values...')
df.dropna(inplace=True)

# Extract necessary features
print('Extracting necessary features...')
df = df[['banknifty', 'nifty', 'tte']]

# Filter data to include only observations within Indian market trading hours (9:15 AM to 3:30 PM)
print('Filtering data for Indian market trading hours...')
df['time'] = pd.to_datetime(df.index, unit='ms').time
start_time = pd.Timestamp('09:15:00').time()
end_time = pd.Timestamp('15:30:00').time()
df = df[df['time'].between(start_time, end_time)]

# Data preprocessing complete
print('Data preprocessing complete.')
print(df.head())

# Logical Data Modeling
# We have data for Bank Nifty and Nifty, along with the time to expiry (TTE) of options.
# We filter the data to include only observations within Indian market trading hours.

# Calculate Moving Averages
window_size = 50  # Adjust window size as needed
df['BankNifty_MA'] = df['banknifty'].rolling(window=window_size).mean()
df['Nifty_MA'] = df['nifty'].rolling(window=window_size).mean()

# Visualization of Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['banknifty'], label='Bank Nifty IV')
plt.plot(df.index, df['BankNifty_MA'], label='Bank Nifty Moving Average ({} periods)'.format(window_size))
plt.plot(df.index, df['nifty'], label='Nifty IV')
plt.plot(df.index, df['Nifty_MA'], label='Nifty Moving Average ({} periods)'.format(window_size))
plt.title('Moving Averages of Bank Nifty and Nifty')
plt.xlabel('Time')
plt.ylabel('Implied Volatility')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
print('Loading dataset...')
df = pd.read_parquet('data.parquet')

# Data Visualization - Implied Volatilities
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['banknifty'], label='Bank Nifty IV')
plt.plot(df.index, df['nifty'], label='Nifty IV')
plt.title('Implied Volatilities of Bank Nifty and Nifty')
plt.xlabel('Time')
plt.ylabel('Implied Volatility')
plt.legend()
plt.show()

# Calculate the spread
df['spread'] = df['banknifty'] - df['nifty']

# Compute z-score of the spread
mean_spread = df['spread'].mean()
std_spread = df['spread'].std()
df['z_score'] = (df['spread'] - mean_spread) / std_spread

# Define entry and exit signals
entry_threshold = 1  # Z-score threshold for entry
exit_threshold = 0  # Z-score threshold for exit
df['position'] = np.where(df['z_score'] > entry_threshold, 1, 0)  # 1 for long position, 0 for no position
df['position'] = np.where(df['z_score'] < -entry_threshold, -1, df['position'])  # -1 for short position
df['position'] = np.where((df['z_score'] > -exit_threshold) & (df['z_score'] < exit_threshold), 0, df['position'])  # Close position when z-score crosses mean

# Calculate P/L based on the provided formula
df['P/L'] = df['spread'] * (df['tte'] ** 0.7)

# Output visualization - Z-score and Entry/Exit Thresholds with Symbols
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['z_score'], label='Z-Score')
plt.axhline(y=entry_threshold, color='r', linestyle='--', label='Entry Threshold')
plt.axhline(y=-entry_threshold, color='r', linestyle='--')
plt.axhline(y=exit_threshold, color='g', linestyle='--', label='Exit Threshold')
plt.axhline(y=-exit_threshold, color='g', linestyle='--')
plt.title('Z-Score and Entry/Exit Thresholds with Symbols')
plt.xlabel('Time')
plt.ylabel('Z-Score')
plt.legend()

# Plot symbols for entry and exit signals
entry_points = df[df['position'] == 1].index
exit_points = df[df['position'] == -1].index
plt.scatter(entry_points, df.loc[entry_points, 'z_score'], color='green', marker='^', label='Entry Signal')
plt.scatter(exit_points, df.loc[exit_points, 'z_score'], color='red', marker='v', label='Exit Signal')

plt.legend()
plt.show()

# Print the dataframe
print(df)

# Calculate cumulative P/L
df['cumulative_PL'] = df['P/L'].cumsum()

# Visualization - Cumulative P/L over time
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['cumulative_PL'], color='purple', label='Cumulative P/L')
plt.title('Cumulative Profit/Loss Over Time')
plt.xlabel('Time')
plt.ylabel('Cumulative P/L')
plt.legend()
plt.grid(True)
plt.show()

# Calculate Sharpe Ratio
daily_returns = df['cumulative_PL'].diff()  # Daily returns
sharpe_ratio = np.mean(daily_returns) / np.std(daily_returns)

# Calculate Drawdown
df['peak'] = df['cumulative_PL'].cummax()
df['drawdown'] = (df['cumulative_PL'] - df['peak']) / df['peak']
max_drawdown = df['drawdown'].min()

# Print P/L, Sharpe Ratio, and Drawdown
print('P/L:', df['cumulative_PL'].iloc[-1])
print('Sharpe Ratio:', sharpe_ratio)
print('Max Drawdown:', max_drawdown)
```
